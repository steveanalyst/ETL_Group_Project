```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
google_file = "C:/Users/Teresa Barajas/Anaconda/Anaconda3/envs/ETL/google_job_skills.csv"
google_df = pd.read_csv(google_file)
google_df.head()
```

```python
cleaned_google_df = google_df.drop_duplicates(['Title'])
cleaned_google_df.dropna()
transformed_google_df = cleaned_google_df[["Company", "Title", "Category", "Responsibilities", "Minimum Qualifications", "Preferred Qualifications"]]
transformed_google_df = transformed_google_df.rename(columns={"Minimum Qualifications":"Minimum_Qualifications", "Preferred Qualifications":"Preferred_Qualifications"})
transformed_google_df.head()
```

```python
glassdoor_file = "C:/Users/Teresa Barajas/Anaconda/Anaconda3/envs/ETL/employee_reviews.csv"
glassdoor_df = pd.read_csv(glassdoor_file)
glassdoor_df.head()
```

```python
glassdoor_google_df = glassdoor_df.loc[glassdoor_df["company"]=="google",:]
glassdoor_google_df 
```

```python
new = glassdoor_google_df["job-title"].str.split("- ", n = 1, expand = True) 
  
# making seperate column from new data frame 
glassdoor_google_df["job_title_cleaned"]= new[1] 
  
```

```python
cleaned_glassdoor_google_df = glassdoor_google_df[["company", "job_title_cleaned", "overall-ratings", "work-balance-stars"]]
cleaned_glassdoor_google_df = cleaned_glassdoor_google_df.rename(columns = {"overall-ratings":"overall_ratings", "work-balance-stars":"work_balance_stars"})
cleaned_glassdoor_google_df = cleaned_glassdoor_google_df [cleaned_glassdoor_google_df.work_balance_stars!="none"]

cleaned_glassdoor_google_df.head()
```
