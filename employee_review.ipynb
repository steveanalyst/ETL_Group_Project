```python
# dependencies
import os
import pandas as pd
```

```python
# load csv data file
employeereview="/Users/wheat/ETL_Group_Project/employee_reviews.csv"
employeereview_df = pd.read_csv(employeereview)
employeereview_df.head()
```

```python
# pull all needed columns
new_employee_df = employeereview_df[['company', 'job-title','overall-ratings','work-balance-stars',
                                     'culture-values-stars','carrer-opportunities-stars','comp-benefit-stars',
                                     'senior-mangemnet-stars','helpful-count']].copy()
new_employee_df
```

```python
# Select data for google and amazon
array = ['google', 'amazon']
new_df=new_employee_df.loc[new_employee_df['company'].isin(array)]
new_df
```

```python
clean1=new_df['job-title'].replace({'Current Employee - ':''}, regex=True)
clean1
```

```python
clean2=clean1.replace({'Former Employee - ':''}, regex=True)
clean2
```

```python
new_df['job-title']=clean2
new_df
```

```python
final_df = new_df[new_df['job-title'] !='Anonymous Employee']
```
